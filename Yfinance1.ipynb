```python
stocks = ['btc', 'eth', 'neo','eos', 'dot', 'bnb', 'ada','xtz']

df3 = pd.DataFrame()

for stock in stocks:
    df3[stocks] = yf.Ticker(stock).history(period = "max").Close
```
